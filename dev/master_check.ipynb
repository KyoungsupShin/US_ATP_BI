```python
smr = SAP_Master_Reset()
smr.atp_batch()
```

```python
view_names = [['ATP_OUTBOUND' , 'OUT_BOUND_HISTORY']
                    , ['ATP_ONHAND', 'ON_HAND_HISTORY']
                    , ['ATP_INBOUND_DOMESTIC', 'INBOUND_DOMESTIC_HISTORY']
                    , ['ATP_INBOUND_OVERSEA', 'INBOUND_OVERSEA_HISTORY']
                    , ['ATP_DELIVERY_STATUS_BI', 'DELIVERY_STATUS_HISTORY']]

# for name in view_names:
name = view_names[0]
smr.connect_azuredb()
smr.df = smr.fetch_data('select * from {}'.format(name[0]))
```

```python
join_check_column_list = ['Product_Name', 'WH_Name']                
join_check_column_list_upper = list(map(str.upper,join_check_column_list))
nullcheck_cols = [col for col in smr.df.columns if col.upper() in join_check_column_list_upper]
nullcheck_total_result = []
if len(nullcheck_cols) > 0:
    for ncols in nullcheck_cols:
        nullcheck_result = smr.df[ncols].isnull().sum() + len(smr.df[smr.df[ncols] == ''])
        if nullcheck_result > 0:
            nullcheck_total_result.append([ncols, nullcheck_result])

if len(nullcheck_total_result) > 0:
    nullcheck_total_result = ['{} : {} ROWS MISSED'.format(i[0], i[1]) for i in nullcheck_total_result]
    nullcheck_total_result = ' <br> '.join(nullcheck_total_result)
    
    print(nullcheck_total_result)
```
