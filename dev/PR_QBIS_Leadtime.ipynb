```python
import sys
sys.path.append('../utils')
from utils import * 

sql = '''
select * from PR_QBIS_INBOUND_TRACKING
order by FactoryPO_NUM, Item_Code, InvoiceNo, Updated_Date;
'''
```

```python
eu = Email_Utils()
eu.connect_azuredb()
df = eu.fetch_data(sql = sql)
```

```python
all_combination = df[df['Data_Type'] != 'PR'][['FactoryPO_Num','Item_Code', 'InvoiceNo']].drop_duplicates()
```

```python
result = []
for i in all_combination.values.tolist()[10:]:
    df_filter = df[(df['FactoryPO_Num'] == i[0]) & (df['Item_Code'] == i[1])]
    if len(df_filter)>1:
        df_filter2 = df[(df['FactoryPO_Num'] == i[0]) & (df['Item_Code'] == i[1]) & (df['InvoiceNo'] == i[2])]
#         df_filter2['LEAD_TIME'] = df_filter2['ATD'] - df_filter[df_filter['Data_Type'] == 'PR'].Updated_Date.values[0]
        pr_date = df_filter[df_filter['Data_Type'] == 'PR']
        atd_date = df_filter2[df_filter2['STATUS'] == 'Onboard']
        if len(pr_date) > 0:
            if len(atd_date) > 0:
                leadtime = (atd_date['ATD'].values[0] - pr_date['Updated_Date'].values[0]).days
                print(i[0], i[1], i[2], atd_date['ATD'].values[0], pr_date['Updated_Date'].values[0], leadtime)
                result.append([i[0], i[1], i[2], atd_date['ATD'].values[0], pr_date['Updated_Date'].values[0], leadtime])
```

```python
df_result = pd.DataFrame(result, columns = ['FactoryPO', 'Item_Code','InvoiceNo', 'ATD', 'PR_Date', 'LeadTime'])
```
