```python
import sys
sys.path.append('../utils')
from utils import * 
```

```python
sql = '''
SELECT FactoryPO_Num, MAX(Dates) AS Dates, Item_Code , SUM(PCS) AS PCS FROM (
	select 
		CASE WHEN LEN(FactoryPO_Num) <= 10 THEN LEFT(FactoryPO_Num ,8) 
		ELSE FactoryPO_Num
		END AS FactoryPO_Num 
		, Dates
		, LEFT(Item_Code,8) AS Item_Code
		, PCS 
	from PURCHASE_ORDER 
) as t1 
GROUP BY FactoryPO_Num, Item_Code 
HAVING MAX(Dates) >= '2022-03-01' '''
```

```python
sql2 = '''
SELECT 
	*
	, SUM(QTY) OVER (PARTITION BY PoNo, itemCode, UPDATED_DATE) AS QTY_DAY
	FROM (select PoNo , itemCode, LOWER(STATUS) AS STATUS, SUM(QTY) AS QTY, Updated_Date   
from DELIVERY_STATUS ds 
WHERE Status in ('Booking', 'Booked', 'cy', 'Onboard') 
group by PoNo , itemCode, STATUS, Updated_Date
) as t1 
order by Updated_Date

'''
```

```python
eu = Email_Utils()
eu.connect_azuredb()
df_pr = eu.fetch_data(sql = sql)
df_qbis = eu.fetch_data(sql = sql2)
```

```python
def get_scm_data(PoNo, status):
    global df_qbis
    global lastest_inbound_qty
    
    status_first_date = df_qbis[(df_qbis['PoNo'] == PoNo) & (df_qbis['STATUS'] == status)  & (df_qbis['itemCode'] == itemCode)].Updated_Date.min()
    status_last_date = df_qbis[(df_qbis['PoNo'] == PoNo) & (df_qbis['STATUS'] == status)  & (df_qbis['itemCode'] == itemCode)].Updated_Date.max()
    if status == 'onboard':
        df_tmp = df_qbis[(df_qbis['QTY_DAY'] == lastest_inbound_qty) & (df_qbis['PoNo'] == PoNo) & (df_qbis['STATUS'] == status)  & (df_qbis['itemCode'] == itemCode)]
        df_tmp = df_tmp.copy()
        df_tmp = df_tmp.drop_duplicates(['PoNo', 'STATUS', 'itemCode', 'QTY', 'QTY_DAY'], keep = 'first')
        status_last_date = df_tmp.Updated_Date.max()
    return status, str(status_first_date), str(status_last_date)

data_traces = []
df_result1 = pd.DataFrame()
```

```python
# PR itemCode가 없는 경우 ?, Qbis에는 있음. 
# PR date추가!!!
```

```python
PoNos = df_pr.FactoryPO_Num.unique()

for PoNo in PoNos:
    print(PoNo)
    itemCodeList = df_qbis[df_qbis['PoNo'] == PoNo].itemCode.unique()
    
    for itemCode in itemCodeList:
        Status = df_qbis[(df_qbis['PoNo'] == PoNo) & (df_qbis['itemCode'] == itemCode)].STATUS.unique()

        # PoNo, PR Qty, QBis Qty
        data_trace = []

        PR_Qty = df_pr[(df_pr['FactoryPO_Num'] ==PoNo) & (df_pr['Item_Code'] == itemCode)].PCS.sum()
        
        if len(df_pr[(df_pr['FactoryPO_Num'] ==PoNo) & (df_pr['Item_Code'] == itemCode)]) >= 1:
            if len(Status) >= 1:
                lastest_inbound_qty = df_qbis[(df_qbis['PoNo'] == PoNo) & (df_qbis['itemCode'] == itemCode)].QTY_DAY.iloc[-1]
                print(lastest_inbound_qty, PR_Qty)
                inbound_progress = round(lastest_inbound_qty/PR_Qty,3)

                for status in Status:
                    status, status_first_date, status_last_date = get_scm_data(PoNo,status)

                    data_trace.append([PoNo, itemCode, PR_Qty, lastest_inbound_qty, inbound_progress, status, status_first_date, status_last_date])

                data_traces.append(data_trace)

                df_result = pd.DataFrame(data_trace, columns = ['PoNo', 'Item_Code','PR_QTY','Inbound_QTY','Inbound_Progress', 'STATUS', 'First_Date', 'Last_Date']).sort_values(['First_Date', 'Last_Date'])
                df_result1 = pd.concat([df_result, df_result1])
            
```
